```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_set= pd.read_csv("/content/drive/MyDrive/dataset/Salary_Data.csv")
```

```python
print(data_set)
```

```python
plt.scatter(data_set["YearsExperience"],data_set["Salary"])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x=data_set[["YearsExperience"]]
y=data_set[["Salary"]]
print(type(x),type(y))
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(xtrain,ytrain)
```

```python
model.predict([[10]])
```

```python
num=float(input("Enter years of exp"))
arr=np.arr([[num]])
print("Predicted salry=",model.predict(arr))
```
